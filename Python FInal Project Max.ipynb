```python

import csv

def create_dictionary_from_csv(file_path, key_column, value_column):
    data = {}
    with open(file_path, 'r') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            #print(row)
            key= row[key_column]
            value = row[value_column]
            value=int(value)
            data[key] = value
    return data
```

```python
file_path = "SPY.csv"
key_column = 'Date'
value_column = "Volume"

Date_Volume = create_dictionary_from_csv(file_path, key_column,value_column)
print(Date_Volume )
```

```python
res = len(Date_Volume)
print("The length of the dictionary is: ", res)
```

# Biggest and smallest values


```python
print(max(Date_Volume.values()))
print(min(Date_Volume,key=Date_Volume.get))
```

```python
print(min(Date_Volume.values()))

print(min(Date_Volume,key=Date_Volume.get))
```
